# Frequency of Retweets



```python
allRetweets={}
for userId in maxRetweeters:
    user=api.get_user(userId)
    tweets=[]
    for tweet in tweepy.Cursor(api.user_timeline,id = userId).items(1000):
        if tweet.text.startswith("RT"):
            tweets.append(tweet.created_at)
    if(len(tweets)>0):
        allRetweets[userId]=tweets
    print(user.screen_name)
```

```python
allRetweets
```

```python
#function to segregate retweets monthly
def freqForEachUser(userId,tweets):
    days={}
    months=["Jan","Feb","Mar","April","May","June","July","Aug","Sep","Oct","Nov","Dec"]
    count=1
    flah=""
    for i in tweets:
        value = months[i.month-1]+" : " + str(i.year)
        if(flah==""):
            flah=value

        if(flah==value):
            count+=1
        else:
            days[flah]=count
            count=1
            flah = value
    days[flah]=count-1
    return days
```

```python
#avegare retweets per month
avgRTPerMonth={}
for userId in allRetweets.keys():
    tweets=allRetweets[userId]
    days=freqForEachUser(userId,tweets)
#     print(userId,days,len(tweets))
    totalRetweets=0
    for key in days.keys():
        totalRetweets+=days[key]
    avg=totalRetweets/len(days)
    avgRTPerMonth[api.get_user(userId).screen_name]=avg

avgRTPerMonth
```

```python
plt.figure(figsize=(20, 10))
plt.bar(avgRTPerMonth.keys(), avgRTPerMonth.values())
plt.title(" Frequency of Retweets")
plt.ylabel('Average retweets per Month')
plt.xlabel('User Name')
plt.show()
```

# Number of different handles which the top retweeters have retweeted
All the top 10 retweeters of @UnderArmour also have retweeted tweets of many others accounts to. This block of code is to find all these accounts.

```python

for userId in maxRetweeters:
    user=api.get_user(userId)
    print(user.screen_name)
    tweets={}
    for tweet in tweepy.Cursor(api.user_timeline,id = userId).items(1000):
        if tweet.text.startswith("RT"):
            text=[]
            for i in tweet.text:
                if i==':':
                    break
                text.append(i)
            text=''.join(text)
            handle=text.split('@')[1]
            if handle in tweets.keys():
                tweets[handle]+=1
            else:
                tweets[handle]=1
            
        
    if(len(tweets)>0):
        allRetweets[userId]=tweets
    
```
