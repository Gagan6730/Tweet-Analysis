```python
lowerTextTweets=[]
for tweet in tweets_text:
    tweet=re.sub('[,\.!?]', '', tweet)
    tweet=re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','',tweet)
    words=tweet.split()
    withoutStop=[i for i in words if i.lower() not in stopwords.words('english')]
    result=[]
    stemmer= PorterStemmer()
    for word in withoutStop:
        result.append(stemmer.stem(WordNetLemmatizer().lemmatize(word, pos='v')))
    tweet=' '.join(result)
    tweet=' '.join(withoutStop)
    lowerTextTweets.append(tweet.lower())
```

```python
def preProcess(tweet):
    tweet=re.sub('[,\.!?]', '', tweet)
    tweet=re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+','',tweet)
    words=tweet.split()
    withoutStop=[i for i in words if i.lower() not in stopwords.words('english')]
    
#     result=[]
#     stemmer= PorterStemmer()
#     for word in withoutStop:
#         result.append(stemmer.stem(WordNetLemmatizer().lemmatize(word, pos='v')))
#     tweet=' '.join(result)
    tweet=' '.join(withoutStop)
    return tweet
    
```

```python
preProcessedTweets_list=[tweet.split() for tweet in lowerTextTweets]
preProcessedTweets_list
```

```python
dictionary = gensim.corpora.Dictionary(preProcessedTweets_list)
```

```python
bagOfWords=[dictionary.doc2bow(l) for l in preProcessedTweets_list]
```

```python
bagOfWords
```

### LDA model with bag of words(vector)

```python
lda_bagOfWords=gensim.models.LdaMulticore(bagOfWords, num_topics=5, id2word=dictionary, passes=10, workers=2)
```

```python
for idx, topic in lda_bagOfWords.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

### LDA model using TFIDF(term frequency per document)

```python
from gensim import corpora, models
```

```python
tfidf = models.TfidfModel(bagOfWords)
```

```python
bow_tfidf=tfidf[bagOfWords]
```

```python
lda_tfidf=gensim.models.LdaMulticore(bow_tfidf, num_topics=5, id2word=dictionary, passes=10, workers=2)
```

```python
for idx, topic in lda_tfidf.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
visual_lda_tfidf= pyLDAvis.gensim.prepare(lda_tfidf, bagOfWords, dictionary)
```
